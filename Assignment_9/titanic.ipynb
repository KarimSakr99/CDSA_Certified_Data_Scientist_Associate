```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(font_scale=1.2)
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.head(2)
```

```python
fig,axes=plt.subplots(1,2,figsize=(18,6))

sns.histplot(df['age'],bins=30,ax=axes[0])
sns.kdeplot(df['age'],ax=axes[1])
```

From the above figure, It is cocluded that the majority of people on the ship are youth between 18 and 40 years old

```python
fig,axes=plt.subplots(1,2,figsize=(18,6))

sns.histplot(df['fare'],bins=30,ax=axes[0])
sns.kdeplot(df['fare'],ax=axes[1])
```

From the above figure, It is cocluded that the majority of people on the ship have paid a very low and equal amount of money ranging between 0 and 80. but, there are some who paid very large amount of money of 500.

```python
df.corr()
```

```python
sns.jointplot(x='age',y='fare',data=df)
```

```python
sns.jointplot(x='age',y='fare',data=df,kind='hex')
```

From the above figures, It is cocluded that fare has nothing to do with age 

```python
sns.jointplot(x='age',y='fare',data=df,hue='sex')
```

From the above figure, It is cocluded that males and females from different ages pay about the same fare. but, females tend to pay a larger fare in some cases 

```python
sns.jointplot(x='age',y='fare',data=df,hue='embark_town')
```

From the above figure, It is cocluded that most of the passengers are from Southampton and Cherbourg.
* That is enough with the hue parameter and leave it for the categorical plotting

```python
sns.pairplot(df,vars=['age','fare'])
```

```python
sns.heatmap(df[['age','fare']].corr())
```

No correlation between age and fare

```python
df.columns
```

```python
fig,axes=plt.subplots(2,4,figsize=(25,10))
sns.countplot(x='sex',data=df,ax=axes[0,0])
sns.countplot(x='survived',data=df,ax=axes[0,1])
sns.countplot(x='class',data=df,ax=axes[0,2])
sns.countplot(x='who',data=df,ax=axes[0,3])
sns.countplot(x='embark_town',data=df,ax=axes[1,0])
sns.countplot(x='alive',data=df,ax=axes[1,1])
sns.countplot(x='alone',data=df,ax=axes[1,2])
```

From the above figure, It is cocluded that:
* about two thirds of the passengers are male
* also about one third of the passengers survived
* majority of the passengers were poor (class third) and the first and second class are about the same number 
* majority of the passengers were men and there are few children
* majority of the passengers were from southampton
* alive gives us the same information as survived feature
* majority of the passengers were alone by themselves on the ship

```python
df.columns
```

```python
sns.swarmplot(data=df,x='sex',y='age',hue='alive')
```

* Majority of men are between 18 and 40 years old and died 
* Majority of women are between 15 and 38 years old and survived 

```python
sns.boxplot(data=df,x='sex',y='fare',hue='embark_town')
```

* There is a lot of outliers
* passengers from Cherbourg pay larger fare

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.boxplot(x='class',y='age',data=df,ax=axes[0])
sns.boxplot(x='class',y='fare',data=df,ax=axes[1])
```

From class,age graph:
* class First has higher ages than the other classes and so on but there are outliers to this rule

From class,fare graph:
* First class of course pay larger fare and so on
* There are some people in class two and there paid very large fare 

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.stripplot(x='who',y='age',hue='alive',data=df,ax=axes[0])
sns.stripplot(x='who',y='fare',hue='alone',data=df,ax=axes[1])
```

* No correlation in both of the above figures between who,age or who,fare
* More women managed to survive than men and children
* threr are some children alone on the ship

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.violinplot(x='embark_town',y='age',data=df,ax=axes[0])
sns.boxplot(x='embark_town',y='fare',data=df,ax=axes[1])
```

* no relation between town and age
* cherbourg pay larger fare but there are outliers from all towns

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.stripplot(x='alive',y='age',data=df,ax=axes[0])
sns.boxplot(x='alive',y='fare',data=df,ax=axes[1])
```

* no relation between age and alive 
* people who pay larger fare may end up alive

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.stripplot(x='alone',y='age',hue='sex',data=df,ax=axes[0])
sns.boxplot(x='alone',y='fare',data=df,ax=axes[1])
```

* alone passangers are more men and can't be under 15 years old but there is a 5-yaer-old female child 
* alone passangers paid less 

```python
sns.barplot(x='embark_town',y='fare',data=df,estimator=sum)
```

The sum of the fare is the largest from southampton due to their number was the heighest

```python
sns.barplot(x='alive',y='fare',data=df)
```

* average fare of alive passengers is double the average fare of the dead
